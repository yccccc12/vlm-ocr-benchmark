## English

```python
import os
import glob
import subprocess

# --- CONFIGURATION ---
DATASET_TYPE = "handwritten_en"   # Change to "handwritten_en" for English dataset
MODEL_NAME = "monkey_ocr"         # Model identifier for output organization

# --- DYNAMIC PATH SETTINGS ---
INPUT_FOLDER = os.path.join("data", "raw", DATASET_TYPE, "images")
BASE_OUTPUT_DIR = os.path.join("outputs", DATASET_TYPE, MODEL_NAME)

# Optional: choose task ('text', 'formula', 'table') or None for default
TASK = "text"

# Optional: enable page splitting for PDFs
PAGE_SPLIT = False

def run_monkey_ocr(file_path, output_dir, task=None, page_split=False):
    """
    Runs Monkey OCR Pro CLI on a file.

    Args:
        file_path (str): Path to image or PDF.
        output_dir (str): Directory to save output.
        task (str, optional): Single recognition task ('text', 'table', 'formula').
        page_split (bool, optional): Whether to enable page splitting for PDFs.
    """

    file_path = os.path.abspath(file_path)
    output_dir = BASE_OUTPUT_DIR

    # output_dir = os.path.abspath(output_dir)
    # Build command
    cmd = ["python3.10", "parse.py", file_path, "-o", output_dir]

    if page_split:
        cmd.append("-s")

    if task:
        cmd.extend(["-t", task])

    print(f"Running command: {' '.join(cmd)}")

    env = os.environ.copy()
    env["MPLBACKEND"] = "Agg"

    result = subprocess.run(
        cmd,
        cwd="MonkeyOCR",
        env=env,
        capture_output=True,
        text=True,
    )


def run_batch_ocr():
    # Ensure the root results folder exists
    # os.makedirs(BASE_OUTPUT_DIR, exist_ok=True)

    # Gather image and PDF files
    file_extensions = ("*.png", "*.jpg", "*.jpeg", "*.pdf")
    file_list = []
    for ext in file_extensions:
        file_list.extend(glob.glob(os.path.join(INPUT_FOLDER, ext)))

    if not file_list:
        print(f"No files found in: {os.path.abspath(INPUT_FOLDER)}")
        return

    print("="*40)
    print(f"STARTING BATCH PROCESS")
    print(f"Dataset: {DATASET_TYPE}")
    print(f"Model:   {MODEL_NAME}")
    print(f"Found:   {len(file_list)} files")
    print("="*40)

    # Process files
    for i, file_path in enumerate(file_list, 1):
        file_name = os.path.basename(file_path)
        print(f"\n[{i}/{len(file_list)}] Processing: {file_name}")
        output_dir = os.path.join(BASE_OUTPUT_DIR, os.path.splitext(file_name)[0])
        run_monkey_ocr(file_path, output_dir, task=TASK, page_split=PAGE_SPLIT)

    print("\n" + "="*40)
    print(f"Batch processing complete!")
    print(f"Results saved in: {os.path.abspath(BASE_OUTPUT_DIR)}")
    print("="*40)

if __name__ == "__main__":
    run_batch_ocr()
```

```python
import os
import glob
import subprocess

# --- CONFIGURATION ---
DATASET_TYPE = "table_by_level"
MODEL_NAME = "monkey_ocr"

# --- DYNAMIC PATH SETTINGS ---
INPUT_BASE_FOLDER = "table_by_level"
BASE_OUTPUT_DIR = os.path.join("outputs", DATASET_TYPE, MODEL_NAME)

# Optional: choose task ('text', 'formula', 'table') or None for default
TASK = "table"

# Optional: enable page splitting for PDFs
PAGE_SPLIT = False

def run_monkey_ocr(file_path, output_dir, task=None, page_split=False):
    """
    Runs Monkey OCR Pro CLI on a file.

    Args:
        file_path (str): Path to image or PDF.
        output_dir (str): Directory to save output.
        task (str, optional): Single recognition task ('text', 'table', 'formula').
        page_split (bool, optional): Whether to enable page splitting for PDFs.
    """
    file_path = os.path.abspath(file_path)
    output_dir = os.path.abspath(output_dir)

    # Ensure the target directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Build command
    cmd = ["python3.10", "parse.py", file_path, "-o", output_dir]

    if page_split:
        cmd.append("-s")

    if task:
        cmd.extend(["-t", task])

    print(f"Running command: {' '.join(cmd)}")

    env = os.environ.copy()
    env["MPLBACKEND"] = "Agg"

    try:
        result = subprocess.run(
            cmd,
            cwd="MonkeyOCR",
            env=env,
            capture_output=True,
            text=True,
        )
        print(f"✅ Saved → {output_dir}")
    except subprocess.CalledProcessError as e:
        print(f"❌ Error processing {os.path.basename(file_path)}:")
        print(e.stderr)

def run_batch_ocr():
    os.makedirs(BASE_OUTPUT_DIR, exist_ok=True)

    if not os.path.exists(INPUT_BASE_FOLDER):
        print(f"❌ Base input folder not found: {os.path.abspath(INPUT_BASE_FOLDER)}")
        return

    print("="*40)
    print(f"🚀 STARTING MONKEY OCR BATCH PROCESS")
    print(f"Dataset: {DATASET_TYPE}")
    print(f"Model:   {MODEL_NAME}")
    print("="*40)

    # Loop through the level folders (e.g., level_1, level_2)
    for level_name in sorted(os.listdir(INPUT_BASE_FOLDER)):
        level_path = os.path.join(INPUT_BASE_FOLDER, level_name)

        # Skip if it's a file
        if not os.path.isdir(level_path):
            continue

        img_folder = os.path.join(level_path, "img")

        if not os.path.isdir(img_folder):
            print(f"⚠️ Warning: 'img' folder not found in {level_name}, skipping...")
            continue

        # Target output directory for this specific level
        level_output_dir = os.path.join(BASE_OUTPUT_DIR, level_name)

        # Gather files
        file_extensions = ("*.png", "*.jpg", "*.jpeg", "*.pdf")
        file_list = []
        for ext in file_extensions:
            file_list.extend(glob.glob(os.path.join(img_folder, ext)))

        if not file_list:
            print(f"⚠️ No files found in: {img_folder}")
            continue

        print(f"\n{'='*40}")
        print(f"📂 LEVEL: {level_name} | Found {len(file_list)} files")
        print(f"{'='*40}")

        # Process files
        for i, file_path in enumerate(file_list, 1):
            file_name = os.path.basename(file_path)
            print(f"\n[{i}/{len(file_list)}] Processing: {file_name}")

            # Create a dedicated subfolder for each image's output inside the level folder
            img_output_dir = os.path.join(level_output_dir, os.path.splitext(file_name)[0])

            run_monkey_ocr(file_path, img_output_dir, task=TASK, page_split=PAGE_SPLIT)

    print("\n" + "="*40)
    print(f"🎉 Batch processing complete!")
    print(f"Results saved in: {os.path.abspath(BASE_OUTPUT_DIR)}")
    print("="*40)

if __name__ == "__main__":
    run_batch_ocr()
```
